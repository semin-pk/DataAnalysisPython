```python
import numpy as np

data = [2, 2, 2, 2, 4, 4, 4, 4, 8, 8, 8, 8]
print("모집단 평균값 :", np.mean(data))
print("불편 분산 값 :", np.var(data, ddof=1))
print("분산 값 :", np.var(data))
print("불편 분산-표준 편차 :", np.std(data,ddof=1))
print("분산-표준 편차 :", np.std(data))
```

```python
#정규 분포 그리기
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
age = np.arange(start=5, stop=60, step=1)

plt.plot(age, stats.norm.pdf(x=age, loc=34, scale=np.sqrt(12)))
```

```python
# 이항 분포
import scipy as sp
import seaborn as sns

binomial = sp.stats.binom(n=10, p = 0.2)
rvs_binomial = binomial.rvs(size=100)
m = np.arange(0, 10, 1)
sns.displot(rvs_binomial, bins=m, kde=False)
```

```python
# 포아송 분포
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

result_1 = sp.stats.poisson.rvs(mu=2, size=1000)
result_2 = sp.stats.poisson.rvs(mu=10, size=1000)
print(result_2)
fig, ax = plt.subplots(2, 1, figsize=(6, 8))
ax[0].hist(result_1, bins=30, label='강도2')
ax[0].legend()
ax[1].hist(result_2, bins=30, label='강도10')
ax[1].legend()
```

```python
import pandas as pd
from scipy import stats
import scipy as sp

data = pd.read_excel('C:/Users/hisem/Desktop/DataAnalysisPython/4장/data/나이vs평점.xlsx')
data.head(3)
```

```python
data.groupby('나이대')[['평점']].mean()
```

```python
score_1030 = data[data['나이대']==1030]['평점'].tolist()
score_4050 = data[data['나이대']==4050]['평점'].tolist()

sp.stats.ttest_ind(score_1030, score_4050, equal_var=False)
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm 

path = 'C:/Users/hisem/Desktop/DataAnalysisPython/4장/data/'

data = pd.read_excel(path + '나이vs평점.xlsx', sheet_name='4그룹')
data['나이대'].value_counts()
```

```python
data['나이대'] = data['나이대'].astype('str')
anova_model = smf.ols("평점 ~ 나이대", data=data).fit()
print(sm.stats.anova_lm(anova_model, typ=2))
```
